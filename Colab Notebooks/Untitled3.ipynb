```python
import matplotlib.pyplot as plt

# Data
rounds = ['Round1', 'Round2', 'Round3', 'Round4', 'Round5']
energy_remaining = [99997.99999, 99996.99999, 99995.99997, 99994.99997, 99993.99997]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(rounds, energy_remaining, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Round')
plt.ylabel('Energy Remaining')
plt.title('Energy Remaining Over Rounds')

# Display the plot
plt.grid(True)
plt.show()
```
